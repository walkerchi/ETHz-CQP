```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.linspace(-1, 2, 1000)
y = np.zeros(len(x))

a = 1

y[(x<a)&(x>=0)] = 1.

fig, ax = plt.subplots(figsize=(8,3))
ax.plot(x, y, lw=3, label="V(x)", alpha=0.5)
ax.set_xlabel("x", fontsize=18)
ax.set_ylabel("V(x)", fontsize=18)
ax.set_title("1-D scattering potential", fontsize=18)
ax.legend(fontsize=18)
ax.set_ylim(-0.1, 1.4)
# hide the xtickes and y tickes
ax.set_xticks([])
ax.set_yticks([])
# put the "a" at the x=1 position and place a vertical line
ax.axvline(x=a, ymin=0, ymax=1, color='orange', linestyle='--')
ax.text(a, -0.1, "a", fontsize=18)
# put the "0" at the x=0 position and place a vertical line
ax.axvline(x=0, ymin=0, ymax=1, color='orange', linestyle='--')
ax.text(0, -0.1, "0", fontsize=18)
# draw a horizontal line at y=0 and put 0 at left 
ax.text(-1.3, 0, "0", fontsize=18)
ax.axhline(y=0, xmin=0, xmax=1, color='purple', linestyle='--')
# fill the space between x=0 and x=a
ax.fill_between(x, y, where=(x>=0)&(x<a), color='lightblue', alpha=0.5)
fig.savefig("1D-scattering-potential.png", dpi=300)
```

```python
import numpy as np
import matplotlib.pyplot as plt 

x = np.linspace(-1, 2, 1000)
y = np.zeros(len(x))

a = 1

# y[(x<0)|(x>=a)] = 1.
y[(x<a)&(x>=0)] = -1.

fig, ax = plt.subplots(figsize=(8,3))
ax.plot(x, y, lw=3, label="V(x)", alpha=0.5)
ax.set_xlabel("x", fontsize=18)
ax.set_ylabel("V(x)", fontsize=18)
ax.set_title("1-D bound state potential", fontsize=18)
ax.legend(fontsize=18)

# hide the xtickes and y tickes
ax.set_xticks([])
ax.set_yticks([])
# put the "a" at the x=1 position and place a vertical line
ax.axvline(x=a, ymin=0, ymax=1, color='orange', linestyle='--')
ax.text(a, -0.1, "a", fontsize=18)
# put the "0" at the x=0 position and place a vertical line
ax.axvline(x=0, ymin=0, ymax=1, color='orange', linestyle='--')
ax.text(0, -0.1, "0", fontsize=18)
ax.text(-1.3, 0, "0", fontsize=18)
ax.axhline(y=0, xmin=0, xmax=1, color='purple', linestyle='--')
# fill the space between x=0 and x=a
ax.fill_between(x, -np.ones_like(y), y, color='lightblue', alpha=0.5)
fig.savefig("1D-bound-state-potential.png", dpi=300)
```

```python
# draw a parabolic potential
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2, 2, 1000)
y = x**2

fig, ax = plt.subplots(figsize=(8,3))
ax.plot(x, y, lw=3, label="V(x)", alpha=0.5)
ax.set_xlabel("x", fontsize=18)
ax.set_ylabel("V(x)", fontsize=18)
ax.set_title("1-D parabolic potential", fontsize=18)

# hide the xtickes and y tickes
ax.set_xticks([])
ax.set_yticks([])

# put horizontal lines at different y = 1,2,3
ax.axhline(y=1, xmin=0, xmax=1, color='purple', linestyle='--')
ax.text(-1.3, 1, "$\ket 1$", fontsize=18)
ax.axhline(y=2, xmin=0, xmax=1, color='purple', linestyle='--')
ax.text(-1.3, 2, "$\ket 2$", fontsize=18)
ax.axhline(y=3, xmin=0, xmax=1, color='purple', linestyle='--')
ax.text(-1.3, 3, "$\ket 3$", fontsize=18)

# put arrows from y=1 to y=2 and y=2 to y=3 annotated with $\hat a^\dagger$
ax.annotate("", xy=(0, 2), xytext=(0, 1), arrowprops=dict(arrowstyle="<->", color='purple'))
ax.annotate("", xy=(0, 3), xytext=(0, 2), arrowprops=dict(arrowstyle="<->", color='purple'))


plt.show()
```
